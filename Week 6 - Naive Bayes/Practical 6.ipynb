# Social Data Mining 2016 - Practical 6: Know your Domains

This session we are going to focus on applying everything we have learned so far to yet another domain. Before we mostly looked at numerical data generated by some social interactions. We also looked at MNIST, a typical Computer Vision dataset. Today, we are going to work with text data. 

- Information: [task & data](https://archive.ics.uci.edu/ml/datasets/SMS+Spam+Collection)
- Data: [spam](https://raw.githubusercontent.com/tcsai/social-data-mining/master/data/spam.csv), [spam_grams](https://raw.githubusercontent.com/tcsai/social-data-mining/master/data/spam_grams.csv)

Again: please make sure you understand the dataset and the task before beginning. In this practical we will focus on predictive classification.

The goal is to distinguish spam (spam SMSes) from ham (actual SMSes). 

The data has been converted to so-called $n$-grams. For this specific set, the word counts for the top 500 most common words were taken.

#### Tasks

- Inspect both `spam.csv` and `spam_grams.csv`.
- Identify and solve issues with the dataset.
- Manually make train, validation, and test splits.
- Compare $k$-nn, Naive Bayes, and a baseline.
- What could we improve about our input?
- What do you think is the advantage Naive Bayes offers over $k$-NN in this case?
